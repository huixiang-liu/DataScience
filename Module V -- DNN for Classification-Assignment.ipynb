```python
# Compile the DNN

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Fit the DNN with your train data

model.fit(X_train, Y_train, validation_data=(X_test, Y_test), nb_epoch=100, batch_size=5)
```

```python
# Evaluate the model
scores = model.evaluate(X_test, Y_test)
print ("Accuracy: %.2f%%" %(scores[1]*100))
```

## Now we start to tune our hyper-parameter

## First, I'll tune activation

```python
def create_model(activation='relu'):
    model = Sequential()
    model.add(Dense(10, input_dim=8, kernel_initializer='uniform',activation=activation))
    model.add(Dense(6))
    model.add(Dense(1, activation="sigmoid"))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
model = KerasClassifier(build_fn=create_model, epochs=100, batch_size=5, verbose=True)
```

```python
activation = ['softmax', 'softplus', 'softsign', 'relu', 'tanh', 'sigmoid', 'hard_sigmoid', 'linear']
param_grid = dict(activation=activation)
```

```python
grid = GridSearchCV(estimator=model, param_grid=param_grid, scoring='accuracy')
```

```python
result = grid.fit(X_train, y_train)
```

```python
print("Best accuracy: {:.2f} using {result}".format(result.best_score_, result=result.best_params_))
```

```python
activation=result.best_params_['activation']
```

## Next, I'll select initializer

```python
def create_model(kernel_initializer='uniform'):
    model = Sequential()
    model.add(Dense(10, input_dim=8, kernel_initializer=kernel_initializer, activation='relu'))
    model.add(Dense(6, activation='relu'))
    model.add(Dense(1, activation="sigmoid"))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
model = KerasClassifier(build_fn=create_model, epochs=100, batch_size=5, verbose=True)
```

```python
param_grid = {"kernel_initializer":['zeros','ones','random_uniform', 'truncated_normal',
                                   'orthogonal']}
```

```python
grid = GridSearchCV(estimator=model, param_grid=param_grid, scoring='accuracy')
```

```python
result = grid.fit(X_train, y_train)
```

```python
print("Best accuracy: {:.2f} using {result}".format(result.best_score_, result=result.best_params_))
```

```python
kernel_initializer=result.best_params_['kernel_initializer']
```

## Finally, we use tuned hyper-parameter on test dataset

```python
def create_model():
    model = Sequential()
    model.add(Dense(10, input_dim=8, kernel_initializer=kernel_initializer, activation=activation))
    model.add(Dense(6, kernel_initializer=kernel_initializer, activation=activation))
    model.add(Dense(1, activation="sigmoid"))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
final_model = create_model()
final_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
final_model.fit(X_train, y_train, validation_data=(X_valid, y_valid), nb_epoch=100, batch_size=5)
```

## Then, let's see how train set accuracy and test set accuracy changes with the increase of epochs

```python
epochs=[1, 5, 10, 20, 50, 100, 200, 500, 1000]
```

```python
result = []
```

```python
for epoch in epochs:
    final_model.fit(X_train, y_train, epochs=epoch, batch_size=5)
    train_result = final_model.evaluate(X_train, y_train)
    valid_result = final_model.evaluate(X_valid, y_valid)
    test_result = final_model.evaluate(X_test, y_test)
    result.append([train_result[1]*100, valid_result[1]*100, test_result[1]*100])
```

```python
result = pd.DataFrame(result)
result.index=epochs
result.columns=["train", "valid", "test"]
result
```

```python
plt.plot(result.iloc[:,:1], label="train")
plt.plot(result.iloc[:,1:2], label="valid")
plt.plot(result.iloc[:,2:], label="test")
plt.xlabel("num_epochs")
plt.ylabel("accuracy")

plt.legend()
plt.show()
```
